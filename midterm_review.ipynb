## Scope of parameters and arguments

- Function parameters are local variables, meaning they exist only within the function’s scope.
- Modifying a parameter inside a function does not affect the original argument passed from outside — unless the argument is a mutable object (like a list).
- Exception: If the argument is a list, then the parameter refers to the same list object. In this case, modifying the list's contents inside the function (e.g., changing elements) will affect the original list outside the function.
- To avoid confusion, it's best to use different variable names in the main program and inside functions.

### Example
```python
# immutable argument (int)
def change_value(x):
    x = 10
    print("Inside function (x):", x)

a = 5
change_value(a)
print("Outside function (a):", a) # still 5
```
The value of a outside the function stays the same because integers are immutable

```python
# mytable argument (list)
def modify_list(my_list):
    my_list[0] = "changed"
    print("Inside function:", my_list)

original_list = ["original", "unchanged"]
modify_list(original_list)
print("Outside function:", original_list)  # First element is changed!
```
The contents of original_list change because lists are mutable, and both original_list and my_list point to the same object in memory.

### What are other ways to copy independent lists?

- L2 = L1[:]
- L2 = L1 + []
- L2 = L1 * 1
- L2 = list(L1)

These create a new list object that contains the same elements as the origninal, but not a reference to the original list itself. 

### L2 = L1[:] - Slicing
- uses list slicing to copy all elements from L2 into a new list L2
- basically, "copy from the first to the last element of L1"
```python
L1 = [1, 2, 3]
L2 = L1[:]   # Creates a new list
print(L2)    # [1, 2, 3]

# new list created. changes to L2 will NOT affect L1
```
### L2 = L1 + [] - Concatenation with empty list
- this adds an empty list to L1, which forces a new list object to be created as the result of the additon
```python
L1 = [1, 2, 3]
L2 = L1 + []  # New list created through addition

# L2 is a new list with the same elements
```

### L2 = L1 * 1 - Multiplication by 1
- multiplies the list by 1, which replicates the list one time - returning a new list with the same contents
```python
L1 = [1, 2, 3]
L2 = L1 * 1   # New list with same elements

# also creates a new list object, even though it looks strange at fist...
```

### L2 = list(L1) - Using the list() Constructor
- converts L1 into a new list by passing it to the built-in list() function
```python
L1 = [1, 2, 3]
L2 = list(L1)  # Calls the constructor to make a new list

# very clear and explicit
```

## Important Note
All of these create **shallow copies**. if the list contains nested lists (i.e., a list of lists), the inner lists are **still shared** between the original and the copy

To copy **deeply**, including all nested elements, use:
```python
import copy
L2 = copy.deepcopy(L1)
```

## Set Methods - used to perform common set operations
- .union() combines **all unique elements** from two (or more) sets
- .intersection() returns elements **common** to both sets
- .difference() returns elements that are in the **first set** but **not in the second**
- .symmetric_difference() returns elements that are **in either set, but not in both**

### What would happen if x were a set or a list instead of a tuple?
- **Tuple Case**: Tuples are immutable, so y[2] = ... raises a TypeError. since y references the same tuple as x, and tuples can't be modified, the code results in an error.
- **Set Case**: Sets are **not indexable**, so x[2] or y[2] would raise a TypeError. Even attempting to access x[2] would fail, regardless of assignment.
- **List Case**: If x were a list (e.g., x = [5, 6, 7, 8, 9], then:

  ```python
  y = x
  y[2] = x[0] + x[1] # y[2] = 5 + 6 = 11
  print(x[2])        # outputs 11

  # Because lists are mutable and y = x creates a reference, modifying y also modifies x
  ```

  <summary><strong>✅ Correct Answer</strong></summary>
  <pre><code>{"a": 1, "b": 1, "c": 1}</code></pre>
  <ul>
    <li>All keys are unique and immutable (str)</li>
    <li>Key-value pairs are separated by commas</li>
    <li>Syntax is correct: {key: value, key: value, ...}</li>
  </ul>
</details>

<details> 
  <summary><strong>ℹ️ Explanation for Each Option</strong></summary>

  <strong>A. {"a":1, "b":2, "b":3}</strong>
  <ul>
    <li>Duplicate key: "b" is defined twice</li>
    <li>Python allows this at runtime, but only the last occurrence survives</li>
    <li>It's valid syntax, but not recommended when you're aiming for clarity or uniqueness in keys</li>
    <li>This might be considered a trick question on the exam...</li>
  </ul>

  <strong>B. {"a":1, "b":1, "c":1}</strong>
  <ul>
    <li>Correct structure — all keys are unique and valid</li>
  </ul>

  <strong>C. {"a":1, ["a", "b"]:2}</strong>
  <ul>
    <li>❌ Invalid key type: <code>["a", "b"]</code> is a list (mutable)</li>
    <li>Dictionary keys must be hashable and immutable — strings, numbers, or tuples (without mutable elements)</li>
    <li>Will raise: <code>TypeError: unhashable type: 'list'</code></li>
  </ul>

  <strong>D. {"a":1; "b":1; "c":1}</strong>
  <ul>
    <li>❌ Invalid syntax: uses semicolons instead of commas</li>
    <li>Python dictionaries require commas between key-value pairs</li>
    <li>Will raise: <code>SyntaxError: invalid syntax</code></li>
  </ul>
</details>

<details> 
  <summary><strong>💡 Quick Tips: Valid Dictionary Rules</strong></summary>
  <ul>
    <li>✅ Keys must be immutable: <code>str</code>, <code>int</code>, <code>float</code>, <code>tuple</code> (no lists inside)</li>
    <li>❌ Keys cannot be mutable types like <code>list</code>, <code>dict</code>, or <code>set</code></li>
    <li>✅ Keys must be unique</li>
    <li>✅ Use commas, not semicolons</li>
    <li>✅ Format: <code>{key1: value1, key2: value2}</code></li>
  </ul>
</details>

<details>
    <summary>📄 Python File Modes Table</summary>

| Mode   | Description                                | Reads? | Writes? | Creates if Not Exists? | Overwrites? | Appends? |
|--------|--------------------------------------------|--------|---------|-------------------------|-------------|----------|
| `"r"`  | Read-only (default)                        | ✅ Yes | ❌ No   | ❌ No                  | ❌ No        | ❌ No     |
| `"w"`  | Write-only, truncates existing content     | ❌ No  | ✅ Yes  | ✅ Yes                 | ✅ Yes       | ❌ No     |
| `"a"`  | Write-only, appends to the end             | ❌ No  | ✅ Yes  | ✅ Yes                 | ❌ No        | ✅ Yes     |
| `"r+"` | Read & write (must exist)                  | ✅ Yes | ✅ Yes  | ❌ No                  | ✅ Yes (if writing from start) | ❌ No     |
| `"w+"` | Read & write, truncates file               | ✅ Yes | ✅ Yes  | ✅ Yes                 | ✅ Yes       | ❌ No     |
| `"a+"` | Read & write, appends to the end           | ✅ Yes | ✅ Yes  | ✅ Yes                 | ❌ No        | ✅ Yes     |
| `"x"`  | Create and write, fails if file exists     | ❌ No  | ✅ Yes  | ✅ Yes (only if new)   | ❌ No (throws error if exists) | ❌ No     |

---

<strong>🧠 Quick Tips</strong>

- **Use `"r"`** when you're only reading an existing file.
- **Use `"w"`** when you want to overwrite or create a new file.
- **Use `"a"`** when you want to keep existing content and add more.
- **Use `+` modes** when you need both **read and write** access.
- **Use `"x"`** when you want to ensure you're creating a **new file only** (and throw an error if it already exists).
